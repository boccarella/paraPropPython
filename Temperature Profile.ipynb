```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
def TemperatureProfile(z):
    Tb = 270
    Ts = 80
    b = 40e3
    h = b / (np.log(Tb/Ts))
    T = Ts * np.exp(z/h)
    return T
z = np.linspace(0, 1000)
    
```

```python
pl.figure(figsize=(10,8))
pl.plot(z, TemperatureProfile(z), label='$T(z) = T_s \cdot exp(z/h)$')
pl.xlabel('Depth [m]', fontsize=22)
pl.ylabel('Temperature [K]', fontsize=22)
pl.xticks(fontsize=20)
pl.yticks(fontsize=20)
pl.legend(shadow=True, loc='upper left', fontsize=18)
pl.ylim(80, 83)
pl.xlim(0,1100)
a = np.where(z==1000)
pl.axhline(TemperatureProfile(z)[a], c = 'r', linestyle='--')
pl.axvline(1000, c='r', linestyle='--')
print(TemperatureProfile(z)[a])
pl.text(630, 82.6, '$T(1000\, m) = 82.47 \,K$', fontsize=18)
pl.scatter(1000, TemperatureProfile(z)[a], c='b', marker='o', s=80)
```
